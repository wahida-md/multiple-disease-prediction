```python
heart_data = pd.read_csv("heart.csv")
heart_data.head()
```

```python
X = heart_data.drop(columns='target', axis=1)
Y = heart_data['target']

print(X.head())
print(Y.head())
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=2
)

print("Data split successful")
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X_train, Y_train)

print("Model trained successfully")
```

```python
from sklearn.metrics import accuracy_score

X_test_prediction = model.predict(X_test)
accuracy = accuracy_score(Y_test, X_test_prediction)

print("Accuracy:", accuracy)
```

```python
import pickle

pickle.dump(model, open('heart_disease_model.sav', 'wb'))

print("Model saved successfully")
```

```python
diabetes_data = pd.read_csv("diabetes.csv")
diabetes_data.head()
```

```python
X = diabetes_data.drop(columns='Outcome', axis=1)
Y = diabetes_data['Outcome']

print(X.head())
print(Y.head())
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=2
)

print("Data split successful")
```

```python
from sklearn.svm import SVC

diabetes_model = SVC(kernel='linear')
diabetes_model.fit(X_train, Y_train)

print("Diabetes model trained successfully")
```

```python
from sklearn.metrics import accuracy_score

X_test_prediction = diabetes_model.predict(X_test)
accuracy = accuracy_score(Y_test, X_test_prediction)

print("Diabetes Model Accuracy:", accuracy)
```

```python
import pickle

pickle.dump(diabetes_model, open('diabetes_model.sav', 'wb'))

print("Diabetes model saved successfully")
```

```python
parkinsons_data = pd.read_csv("parkinsons.csv")
parkinsons_data.head()
```

```python
X = parkinsons_data.drop(columns=['name', 'status'], axis=1)
Y = parkinsons_data['status']

print(X.head())
print(Y.head())
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=2
)

print("Parkinson’s data split successful")
```

```python
from sklearn.svm import SVC

parkinsons_model = SVC(kernel='linear')
parkinsons_model.fit(X_train, Y_train)

print("Parkinson’s model trained successfully")
```

```python
from sklearn.metrics import accuracy_score

X_test_prediction = parkinsons_model.predict(X_test)
accuracy = accuracy_score(Y_test, X_test_prediction)

print("Parkinson’s Model Accuracy:", accuracy)
```

```python
import pickle

pickle.dump(parkinsons_model, open('parkinsons_model.sav', 'wb'))

print("Parkinson’s model saved successfully")
```
